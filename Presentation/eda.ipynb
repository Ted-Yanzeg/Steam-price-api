```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#  1. Load data 
df = pd.read_csv("../data/steam_games.csv")  
df['is_free'] = df['price_usd'] == 0.0
```

```python
# 2. Price vs. Year  
df_paid = df[df['price_usd'] > 0]
median_by_year = df_paid.groupby('release_year')['price_usd'].median()

plt.figure(figsize=(10, 5))
plt.plot(median_by_year.index, median_by_year.values, marker='o')
plt.title('Median Price of Paid Games by Release Year')
plt.xlabel('Release Year')
plt.ylabel('Median Price (USD)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
free_vs_paid = df['is_free'].value_counts().rename({True: 'Free', False: 'Paid'})

plt.figure(figsize=(6, 4))
plt.pie(free_vs_paid.values, labels=free_vs_paid.index, autopct='%1.1f%%')
plt.title('Free vs Paid Games Distribution')
plt.tight_layout()
plt.show()
```

```python

df_paid = df[df["price_usd"] >= 0]

# Price vs Total Reviews
plt.figure(figsize=(6, 5))
plt.scatter(df_paid["total_reviews"], df_paid["price_usd"], alpha=0.4)
plt.title("Price vs Total Reviews")
plt.xlabel("Total Reviews")
plt.ylabel("Price (USD)")
plt.grid(True)
plt.tight_layout()
plt.show()
```
